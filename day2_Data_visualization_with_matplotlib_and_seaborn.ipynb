```python
# --------------------------
# Part 1: Matplotlib Basics
# --------------------------
import matplotlib.pyplot as plt

# Sample data
students = ["Alice", "Bob", "Charlie", "David", "Eva"]
scores = [85, 90, 78, 88, 95]

# 1. Line Plot
plt.figure(figsize=(8,5))
plt.plot(students, scores, marker='o', linestyle='-', color='blue')
plt.title("Line Plot of Student Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.grid(True)
plt.show()







```

```python

# 2. Bar Plot
plt.figure(figsize=(8,5))
plt.bar(students, scores, color='orange')
plt.title("Bar Plot of Student Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()

```

```python

# 3. Horizontal Bar Plot
plt.figure(figsize=(8,5))
plt.barh(students, scores, color='green')
plt.title("Horizontal Bar Plot of Student Scores")
plt.xlabel("Scores")
plt.ylabel("Students")
plt.show()
```

```python
# 4. Scatter Plot
ages = [20, 22, 19, 21, 23]
plt.figure(figsize=(8,5))
plt.scatter(ages, scores, color='red', s=100)
plt.title("Scatter Plot: Age vs Scores")
plt.xlabel("Age")
plt.ylabel("Scores")
plt.show()
```

```python


# 5. Histogram
plt.figure(figsize=(8,5))
plt.hist(scores, bins=5, color='purple', edgecolor='black')
plt.title("Histogram of Scores")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.show()
```

```python
# --------------------------
# Part 2: Seaborn Basics
# --------------------------

import seaborn as sns
import pandas as pd

data = {
    "Name": ["Amrita", "Bijay", "Ram", "Darshan", "Eva", "Suresh", "Grace", "Ramesh", "Isha", "Hari"],
    "Age": [20, 22, 19, 21, 23, 20, 22, 21, 20, 22],
    "Score": [85, 90, 78, 88, 95, 82, 91, 87, 80, 89],
    "Gender": ["F", "M", "M", "M", "F", "M", "F", "F", "F", "M"]
}

df = pd.DataFrame(data)
# 1. Scatter Plot with Seaborn
sns.scatterplot(data=df, x="Age", y="Score", hue="Gender", style="Gender",s=100)
plt.title("Seaborn Scatter Plot: Age vs Score by Gender")
plt.show()
```

```python
# 2. Line Plot
sns.lineplot(data=df, x="Name", y="Score", marker="o")
plt.title("Seaborn Line Plot: Scores by Students")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Bar Plot
sns.barplot(data=df, x="Name", y="Score", hue="Gender")
plt.title("Seaborn Bar Plot: Scores by Student & Gender")
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Histogram / Distribution Plot
sns.histplot(df["Score"], bins=5, kde=True, color="skyblue")
plt.title("Distribution of Scores")
plt.show()
```

```python
# 5. Box Plot
sns.boxplot(data=df, x="Gender", y="Score", palette="Set2")
plt.title("Box Plot: Score distribution by Gender")
plt.show()
```

```python
# 6. Pair Plot
sns.pairplot(df, hue="Gender", diag_kind="kde")
plt.show()
```

```python
# --------------------------
# Part 3: Customization & Styling
# --------------------------

# Example: Customized Matplotlib plot
plt.figure(figsize=(8,5))
plt.plot(students, scores, marker='o', linestyle='--', color='magenta', linewidth=2)
plt.title("Customized Line Plot", fontsize=16)
plt.xlabel("Students", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.grid(True, linestyle=':', linewidth=0.8)
plt.show()

# Example: Seaborn style change
sns.set_style("whitegrid")
sns.barplot(data=df, x="Name", y="Score", palette="coolwarm")
plt.title("Seaborn Bar Plot with Whitegrid Style")
plt.xticks(rotation=45)
plt.show()
```
